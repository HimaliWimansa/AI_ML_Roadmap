# 🚀 Iris Model Deployment API

## 📌 Overview
This project deploys a machine learning model using Flask to serve predictions via an API.

## ⚙️ Features
- REST API endpoint for predictions
- Accepts JSON input
- Returns predicted iris species

## 🔌 API Endpoint
POST /predict

### Input Example
{
  "features": [5.1, 3.5, 1.4, 0.2]
}

### Output Example
{
  "prediction": "Iris_verginica"
}

## 🛠️ Tech Stack
- Python
- Flask
- Scikit-learn
- Joblib
